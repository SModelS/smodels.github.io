```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

## Higgsino scenario, BR to Z = 0.5, combined SRs

```python
#with combination
with open('summary.txt','r') as file:
    with open("rvalues_brZ50_comb.dat",'w') as out:
        for line in file.readlines():
            if '#' in line : continue
            x = int(line.split()[0].split('_')[3])
            y = int(line.split()[0].split('_')[4])
            z = float(line.split()[2])
            out.write('{} {} {} \n'.format(x,y,z))
```

```python
dt = np.dtype([('m2', float), ('m1', float), ('r', float)])
allPts = np.genfromtxt('rValues_brZ50.dat', dtype=dt, usecols=range(0,3))
len(allPts)
```

```python
obsLim = np.genfromtxt('obsLimitAtlas_brZ50.dat', dtype=np.dtype([('m2', float), ('m1', float)]), usecols=range(0,2))
len(obsLim)
```

```python
#fig = plt.figure(figsize=(6,4))
fig = plt.figure()

plt.tick_params(which="major", length=5, direction="in", bottom=True, top=True, left=True, right=True)
plt.tick_params(which="minor", axis="x", direction="in", bottom=True, top=True, length=4)
plt.tick_params(which="minor", axis="y", direction="in", bottom=True, top=True, length=4)
plt.tick_params(labelbottom=True, labelleft=True, labeltop=False, labelright=False)

plt.ylim(0, 410)
#plt.grid(alpha=0.6)

plt.scatter(allPts['m2'],allPts['m1'], c=allPts['r'], cmap='coolwarm', s=30, alpha=1, vmin=0.5, vmax=1.5)
cbar = plt.colorbar()
cbar.set_label(r'$r$-value', rotation=90,fontsize=14)

plt.plot(obsLim['m2'],obsLim['m1'], c='k', linewidth=2, label='ATLAS official')
plt.legend(loc='upper right', frameon=True, fontsize=14)

plt.xlabel(r'$m_{\tilde H}$ [GeV]', fontsize=14)
plt.ylabel(r'$m_{\tilde a}$ [GeV]', fontsize=14)
plt.title(r'ATLAS-SUSY-2018-41, BR$(\tilde H\to \tilde a Z)=0.5$, combSR=True',loc='center',fontsize=12)

plt.tight_layout()
plt.show()
fig.savefig('higgsino_BRZ50_comb.png', dpi=150)
```
