```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

## Higgsino scenario, BR to Z = 0.5, combined vs best SR (H~,a~)

```python
dt = np.dtype([('m2', float), ('m1', float), ('r', float)])
combSR = np.genfromtxt('validationMixedBRs/rValues_brZ50_combSR.dat', dtype=dt, usecols=range(0,3))
bestSR = np.genfromtxt('validationMixedBRs/rValues_brZ50_bestSR.dat', dtype=dt, usecols=range(0,3))
[len(bestSR),len(combSR)]
```

```python
obsLim = np.genfromtxt('orig/Obslimiton(H~,a~)B(N1->Za~)=50%.csv', 
                       dtype=np.dtype([('m2', float), ('m1', float)]), 
                       usecols=range(0,2), delimiter=',')
len(obsLim)
```

```python
fig = plt.figure(figsize=(6,4))

plt.tick_params(which="major", length=5, direction="in", bottom=True, top=True, left=True, right=True)
plt.tick_params(which="minor", axis="x", direction="in", bottom=True, top=True, length=4)
plt.tick_params(which="minor", axis="y", direction="in", bottom=True, top=True, length=4)
plt.tick_params(labelbottom=True, labelleft=True, labeltop=False, labelright=False)

plt.ylim(-10, 410)
#plt.grid(alpha=0.6)

plt.scatter(combSR['m2'],combSR['m1'], c=combSR['r'], cmap='jet', s=30, alpha=1, vmin=0, vmax=1.6)
cbar = plt.colorbar()
cbar.set_label(r'$r$-value', rotation=90,fontsize=14)

plt.plot(obsLim['m2'],obsLim['m1'], c='k', linewidth=2, label='ATLAS official')
plt.legend(loc='upper right', frameon=True, fontsize=14)

plt.xlabel(r'$m_{\tilde H}$ [GeV]', fontsize=14)
plt.ylabel(r'$m_{\tilde a}$ [GeV]', fontsize=14)
plt.title(r'ATLAS-SUSY-2018-41, BR$(\tilde\chi^0_1\to \tilde a Z)=0.5$, combSR=True',loc='center',fontsize=12)

plt.tight_layout()
plt.show()
fig.savefig('validationMixedBRs/Higgsino(H~,a~)_BRZ50_comb.png', dpi=150)
```

```python
fig = plt.figure(figsize=(6,4))

plt.tick_params(which="major", length=5, direction="in", bottom=True, top=True, left=True, right=True)
plt.tick_params(which="minor", axis="x", direction="in", bottom=True, top=True, length=4)
plt.tick_params(which="minor", axis="y", direction="in", bottom=True, top=True, length=4)
plt.tick_params(labelbottom=True, labelleft=True, labeltop=False, labelright=False)

plt.ylim(-10, 410)
#plt.grid(alpha=0.6)

plt.scatter(bestSR['m2'],bestSR['m1'], c=bestSR['r'], cmap='jet', s=30, alpha=1, vmin=0, vmax=1.6)
cbar = plt.colorbar()
cbar.set_label(r'$r$-value', rotation=90,fontsize=14)

plt.plot(obsLim['m2'],obsLim['m1'], c='k', linewidth=2, label='ATLAS official')
plt.legend(loc='upper right', frameon=True, fontsize=14)

plt.xlabel(r'$m_{\tilde H}$ [GeV]', fontsize=14)
plt.ylabel(r'$m_{\tilde a}$ [GeV]', fontsize=14)
plt.title(r'ATLAS-SUSY-2018-41, BR$(\tilde\chi^0_1\to \tilde a Z)=0.5$, combSR=False',loc='center',fontsize=12)

plt.tight_layout()
plt.show()
fig.savefig('validationMixedBRs/Higgsino(H~,a~)_BRZ50_best.png', dpi=150)
```

```python
fig = plt.figure(figsize=(6,4.5))

plt.plot(obsLim['m2'],obsLim['m1'], c='k', linewidth=2, 
         label='official ATLAS exclusion')

plt.tricontour(bestSR['m2'],bestSR['m1'],bestSR['r'], levels=[1.], colors=['r'], linestyles=['-'])
plt.tricontour(combSR['m2'],combSR['m1'],combSR['r'], levels=[1.], colors=['r'], linestyles=['--'])
plt.plot([0,1], [0,1], c='r', ls='-', linewidth=2, label='SModelS: best SR')
plt.plot([0,1], [0,1], c='r', ls='--', linewidth=2, label='SModelS: naive SR comb.')

plt.legend()

plt.xlim(200, 1200)
plt.ylim(0, 350)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tick_params(direction='in',length=8, width=1, top=True, bottom=True, left=True, right=True)

plt.xlabel(r'$m_{\tilde H}$ [GeV]',fontsize=14)
plt.ylabel(r'$m_{\tilde a}$ [GeV]',fontsize=14)
plt.title(r'ATLAS-SUSY-2018-41, $pp\,\to\,\tilde H\tilde H$ $(\tilde H=\tilde\chi^\pm_1,\tilde\chi^0_2,\tilde\chi^0_1)$',
          fontsize=12)

plt.text(270,290, r'BR$(\tilde\chi^0_1\,\to\,\tilde aZ)=0.5$', fontsize=12)
plt.text(270,260, r'BR$(\tilde\chi^0_1\,\to\,\tilde ah)=0.5$', fontsize=12)

plt.tight_layout()
plt.show()
fig.savefig('validationMixedBRs/Higgsino(H~,a~)_BRZ50.png', dpi=150)
```
