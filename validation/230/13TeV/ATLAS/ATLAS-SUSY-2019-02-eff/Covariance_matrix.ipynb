```python
import csv
```

```python
obsN = []
expBg = []
bgErr = []
```

```python
with open("orig/TChiWW_yields.csv",'r') as file:
    csvreader = csv.reader(file)
    for row in csvreader:
        if '#' in row[0] : continue
        obsN.append(int(row[1]))
        expBg.append(round(float(row[2]),2))
        bgErr.append(round(float(row[3]),2))
```

```python
bgErr[0]**2
```

```python
x = 1
for err1 in bgErr:
        y = 1
        for err2 in bgErr:
            if x == y :
                z = round(err1**2,2)
                #out.write('{},{},{} \n'.format(x,y,z))
                print(x,y,z,'\n')
            elif (x<=16 and y >16) or (x>16 and y<=16):
                z = 0.0
                #out.write('{},{},{} \n'.format(x,y,z))
                print(x,y,z,'\n')
            else :
                z = round(0.99*err1*err2,2)
                #out.write('{},{},{} \n'.format(x,y,z))
                print(x,y,z,'\n')
            y+=1
        x+=1
```

```python
with open("CovarianceMatrix_nontrivialcorrelation.csv",'w') as out:
    out.write('#: table: created manually for TChiWW efficiencies \n')
    out.write('#: name: Covariance matrix \n')
    out.write('#: description: assuming max (0.99) correlations amongst individual DF SRs and SF SRs and no correlation between DF and SF SRs. The 24 SRs considered are exclusive(binned) = [DF_81_8125, DF_8125_815, DF_815_8175, DF_8175_82, DF_82_8225, DF_8225_825, DF_825_8275, DF_8275_83, DF_83_8325, DF_8325_835, DF_835_8375, DF_8375_84 , DF_84_845, DF_845_85, DF_85_86, DF_86_1, SF_77_775, SF_775_78, SF_78_785, SF_785_79, SF_79_795, SF_795_80, SF_80_81, SF_81_1] \n')
    out.write('#: keyword observables: COVARR \n')
    out.write('#: keyword phrases: SUSY | Supersymmetry \n')
    out.write('#: keyword cmenergies: 13000 \n')
    out.write('#: keyword reactions: P P --> W W INVISIBLE \n')
    out.write('#: LUMINOSITY [fb$^{-1}$],139 \n')
    out.write('"Bin, x","Bin, y",Covariance matrix elements \n')
    x = 1
    for err1 in bgErr:
        y = 1
        for err2 in bgErr:
            if x == y :
                z = round(err1**2,2)
                out.write('{},{},{} \n'.format(x,y,z))
            elif (x<=16 and y >16) or (x>16 and y<=16):
                z = 0.0
                out.write('{},{},{} \n'.format(x,y,z))
            else :
                z = round(0.99*err1*err2,2)
                out.write('{},{},{} \n'.format(x,y,z))
            y+=1
        x+=1
```
